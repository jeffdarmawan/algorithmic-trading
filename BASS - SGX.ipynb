```python
benchmark = "^STI"

start = "2023-01-01"
end = "2024-01-01"
```

```python
def BASS(stock_symbol, benchmark_symbol, start, end):
    df = pd.DataFrame()
    
    # Download ticker5GJ
    df["stock"] = yf.Ticker(stock_symbol).history(start=start,end=end).Close

    # check if data available
    if len(df["stock"]) == 0:
        return None, None, None, None
        
    df["benchmark"] = yf.Ticker(benchmark).history(start=start,end=end).Close

    # Calculate Return
    df["benchmark_returns"] = df["benchmark"].pct_change()
    df["stock_returns"] = df["stock"].pct_change()

    df = df.dropna()
    df.head()

    # Risk (Standard Deviation)
    std_dev = (df["stock_returns"].std()) * 100
    std_dev = round(std_dev, 2)

    if std_dev == 0:
        print("something's wrong. ticker: " + stock_symbol + ". std_dev: " + str(std_dev))
        return None, None, None, None

    # Beta
    cov = df["benchmark_returns"].cov(df["stock_returns"])
    var = df["benchmark_returns"].var()
    beta = cov/var
    beta = round(beta,2)

    # Alpha
    benchmark_yearly_returns = (df["benchmark_returns"].mean()*252)
    stock_yearly_returns = (df["stock_returns"].mean()*252)
    alpha = (stock_yearly_returns - beta * benchmark_yearly_returns) * 100
    alpha = round(alpha, 2)

    # Sharpe (Daily)
    avg_returns = df['stock_returns'].mean()
    std = df['stock_returns'].std()
    daily_SR = avg_returns / std

    # Sharpe (Annual)
    annual_SR = daily_SR * (252 ** 0.5)
    annual_SR = round(annual_SR,2)
    
    
    return beta, alpha, std_dev, annual_SR
```

```python
bass_list = []
final_stocks = []

# iterate ticker to retrieve its BASS and append it into our main dataframe
for ticker in all_stocks:
    bass_ticker = BASS(ticker, benchmark, start, end)

    # ignore annomalies
    # - static stock price
    if bass_ticker[0] == None:
        continue

    bass_list.append(bass_ticker)
    final_stocks.append(ticker)
```

```python
bass_final = pd.DataFrame(bass_list, columns=['beta', 'alpha', 'std_dev', 'sharpe_ratio'], index=final_stocks)
bass_final.index.name = 'stock_code'
bass_final = bass_final.sort_values(by='sharpe_ratio', ascending=False)

bass_final.head(20)
```
