```python
from ta.utils import dropna
from ta.volatility import AverageTrueRange
from ta.trend import SMAIndicator
from ta.momentum import RSIIndicator

import yfinance as yf
```

```python
stock_symbol = 'AAPL'
start = "2023-01-01"
end = "2024-01-01"

sma_window = 5
rsi_window = 14
atr_window = 14
```

```python
# Stock prices
stock_data = yf.Ticker(stock_symbol).history(start=start,end=end)[['Open','High','Low','Close','Volume']]
stock_data 
```

```python
# SMA Indicator
stock_data['sma_5'] = SMAIndicator(close = stock_data.Close, window = sma_window, fillna = False).sma_indicator()
stock_data['sma_5'].head(15)
```

```python
# RSI Indicator
stock_data['rsi_14'] = RSIIndicator(close = stock_data.Close, window = rsi_window, fillna = False).rsi()
stock_data['rsi_14'].head(20)
```

```python
# Average True Range
stock_data['atr_14'] = AverageTrueRange(high = stock_data.High, low = stock_data.Low, close = stock_data.Close, window = atr_window, fillna = False).average_true_range()
stock_data['atr_14'].head(20)
```
